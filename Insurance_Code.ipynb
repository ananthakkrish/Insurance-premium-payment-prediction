# Dropped Variables

```python
#Dropped "VAR1","VAR11","VAR12","VAR15","VAR20","VAR21","VAR23", seeing them being repetitions of other columns
train=train.drop(["VAR1","VAR3","VAR9","VAR11","VAR12","VAR15","VAR20","VAR21","VAR23","VAR27","VAR35","VAR37"], axis=1)
test=test.drop(["VAR1","VAR3","VAR9","VAR11","VAR12","VAR15","VAR20","VAR21","VAR23","VAR27","VAR35","VAR37"], axis=1)
oot=oot.drop(["VAR1","VAR3","VAR9","VAR11","VAR12","VAR15","VAR20","VAR21","VAR23","VAR27","VAR35","VAR37"], axis=1)
```

# Removing Nulls

These variables had less NA values, those rows were removed

```python
#Remove nulls in 'VAR3','VAR5','VAR6','VAR9','VAR14','VAR16','VAR19','VAR22','VAR26','VAR30'
train=train.dropna(subset=['VAR5','VAR6','VAR14','VAR16','VAR19','VAR22','VAR26','VAR30'], axis=0, how='any')
train.shape

oot=oot.dropna(subset=['VAR5','VAR6','VAR14','VAR16','VAR19','VAR22','VAR26','VAR30'], axis=0, how='any')
test=test.dropna(subset=['VAR5','VAR6','VAR14','VAR16','VAR19','VAR22','VAR26','VAR30'], axis=0, how='any')

test1=test.copy
oot1=oot.copy
```

# VAR2 

The persistency percentages were converted to decimals and null values and '-' were imputed with 0


```python

train.VAR2=train["VAR2"].replace('-',0)
train["VAR2"]=train["VAR2"].astype(float)

for col in train.columns:
    for row in train.index:
        if train.loc[row,"VAR2"]>1:
            train.loc[row,"VAR2"]= train.loc[row,"VAR2"]/100
train["VAR2"]=train["VAR2"].fillna(0) 

#test
test.VAR2=test["VAR2"].replace('-',0)
test["VAR2"]=test["VAR2"].astype(float)

for col in test.columns:
    for row in test.index:
        if test.loc[row,"VAR2"]>1:
            test.loc[row,"VAR2"]= test.loc[row,"VAR2"]/100
test["VAR2"]=test["VAR2"].fillna(0)   

#OOT
oot.VAR2=oot["VAR2"].replace('-',0)
oot["VAR2"]=oot["VAR2"].astype(float)

for col in oot.columns:
    for row in oot.index:
        if oot.loc[row,"VAR2"]>1:
            oot.loc[row,"VAR2"]= oot.loc[row,"VAR2"]/100
oot["VAR2"]=oot["VAR2"].fillna(0)   
```

# VAR4

NA values were first filled with 0 and then No was replaced with 1 and yes with 0

```python
#Train
train['VAR4']=train['VAR4'].fillna(0)
train.VAR4=train["VAR4"].replace('N',1)
train.VAR4=train["VAR4"].replace('Y',2)
## 0 - no response 
## 1- no alcohol
## 2 - yes

test['VAR4']=test['VAR4'].fillna(0)
test.VAR4=test["VAR4"].replace('N',1)
test.VAR4=test["VAR4"].replace('Y',2)


oot['VAR4']=oot['VAR4'].fillna(0)
oot.VAR4=oot["VAR4"].replace('N',1)
oot.VAR4=oot["VAR4"].replace('Y',2)
```

# VAR7

```python
train['VAR7'] = np.where(train['VAR7']=='Y',1,0)
test['VAR7'] = np.where(test['VAR7']=='Y',1,0)
oot['VAR7'] = np.where(oot['VAR7']=='Y',1,0)
```

# VAR8 & VAR14

## Bmi has been imputed as per the average age bmi

```python
#VAR8
# bmi outliers treating as 10 to 50 
train = train[~((train['VAR8'] >= 0.1) & (train['VAR8'] < 10))]
train = train[~(train['VAR8'] > 50)]

#VAR14
# age 
train = train[train['VAR14'] >= 0]

# mean calculating
age_bmi_list = train.groupby('VAR14')['VAR8'].mean().to_frame()

# setting index
train = train.set_index('VAR14')

# replace
#Python
#Col 1 = where you want the values replaced
#Col 2 = where you want to take the values from
train["VAR8"].fillna(age_bmi_list["VAR8"], inplace=True)
train = train.reset_index()

#test["VAR8"].fillna(age_bmi_list["VAR8"], inplace=True)
#oot["VAR8"].fillna(age_bmi_list["VAR8"], inplace=True)
```

```python
#VAR8
# bmi outliers treating as 10 to 50 
test = test[~((test['VAR8'] >= 0.1) & (test['VAR8'] < 10))]
test = test[~(test['VAR8'] > 50)]

#VAR14
# age 
test = test[test['VAR14'] >= 0]

# mean calculating
age_bmi_list_test = test.groupby('VAR14')['VAR8'].mean().to_frame()

# setting index
test = test.set_index('VAR14')

# replace
#Python
#Col 1 = where you want the values replaced
#Col 2 = where you want to take the values from
test["VAR8"].fillna(age_bmi_list_test["VAR8"], inplace=True)
test = test.reset_index()
```

```python
#VAR8
# bmi outliers treating as 10 to 50 
oot = oot[~((oot['VAR8'] >= 0.1) & (oot['VAR8'] < 10))]
oot = oot[~(oot['VAR8'] > 50)]

#VAR14
# age 
oot = oot[oot['VAR14'] >= 0]

# mean calculating
age_bmi_list_oot = oot.groupby('VAR14')['VAR8'].mean().to_frame()

# setting index
oot = oot.set_index('VAR14')

# replace
#Python
#Col 1 = where you want the values replaced
#Col 2 = where you want to take the values from
oot["VAR8"].fillna(age_bmi_list_oot["VAR8"], inplace=True)
oot = oot.reset_index()
```
